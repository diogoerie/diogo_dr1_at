```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

knn_res = np.array([0.8575, 0.8583, 0.8584, 0.8580, 0.8585])
regressao_res = np.array([0.8600, 0.8506, 0.8600, 0.8599, 0.8600])

dados_resultados = {
    'Modelo': ['KNN', 'Regressão Logística'],
    'Acurácia Média (%)': [knn_res.mean() * 100, regressao_res.mean() * 100],
    'Desvio Padrão (%)': [knn_res.std() * 100, regressao_res.std() * 100]
}

df_resultados = pd.DataFrame(dados_resultados)
display(df_resultados.round(2))
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(df_resultados['Modelo'],
       df_resultados['Acurácia Média (%)'],
       yerr=df_resultados['Desvio Padrão (%)'],
       capsize=8,
       color=['skyblue', 'lightgreen'])

ax.set_ylabel('Acurácia (%)')
ax.set_title('Comparação entre Modelos – Validação Cruzada')
ax.grid(True, axis='y', linestyle='--', alpha=0.6)

plt.show()
```
